```python
import pandas as pd

gitCloudRepo='https://github.com/PUBPOL-542/D1/raw/main/'
fileName1="king20.xlsx"
fileName2="king19.xlsx"
fileName3="king21.xlsx"
fileName4="king3y.xlsx"

DFexcel1=pd.read_excel(gitCloudRepo + fileName1)
DFexcel2=pd.read_excel(gitCloudRepo + fileName2)
DFexcel3=pd.read_excel(gitCloudRepo + fileName3)
DFexcel4=pd.read_excel(gitCloudRepo + fileName4)
```

```python
(DFexcel1.columns,
DFexcel2.columns,
DFexcel3.columns,
DFexcel4.columns)
```

```python
DFconcat = pd.concat([DFexcel1, DFexcel3, DFexcel3])
DFconcat
```

```python
result = pd.merge(DFconcat, DFexcel4, on="serial")
result.head()
```

```python
result.info()
```

```python
bye=[1,3,4,5,5,9,10,15,16,17]
result.drop(columns=result.columns[bye],inplace=True)
```

```python
result.info()
```

```python
result['hhincome'].replace(9999999,None,inplace=True)
```

```python
result["hhincome"] = pd.to_numeric(result["hhincome"])
```

```python
result.info()
```

```python
result.head()
```

```python
result.rename(columns={'year_x': 'year','statefip': 'state','ownershp_x': 'ownership'}, inplace=True)
```

```python
result.info()
```

```python
result.to_csv('result.csv', index=False)
```
